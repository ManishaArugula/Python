## Exploratory Data Analysis 

#### Analysis questions
1. Determine gender distribution in the dataset
2. Revenue breakdown by gender

```python
plt.figure(figsize=(11,6))
ax = sns.countplot(x='Gender', data =store, palette='Set2')
for data_label in ax.containers:
    ax.bar_label(data_label)
```

```python
revenue_by_gender = store.groupby('Gender', as_index=False)['Amount'].sum() # To make sure the result is in the form of 
#dataframe use the as.index= False indication no column is an index
revenue_by_gender
```

```python
total_revenue = store['Amount'].sum()
total_revenue
```

```python
revenue_by_gender['revenue_percent']= (revenue_by_gender['Amount']/total_revenue)*100
```

```python
revenue_by_gender
```

```python
plt.figure(figsize=(10, 5)) 
slice_explode = (0.03, 0) # We have only two slices, hence we give only two parameters
plt.pie(revenue_by_gender['revenue_percent'],autopct='%.0f%%', labels=revenue_by_gender['Gender'], 
        explode=slice_explode,shadow=True, colors=['#B2DFDB','#F28C8C'])
plt.show()
```

#### Women account for the highest shopping volume, generating 70% of revenue.

### Analysis based on Age
1. Distribution of the age groups along with gender in the dataset
2. Which age group contributes the highest to store's revenue

```python
##sorting by the number of count values associated with each age group
age_counts = store['Age Group'].value_counts().index
age_counts
```

```python
#sorting based on just indexs
store['Age Group'].sort_index().unique()
```

```python
##Distibution of age groups and gender in the dataset
##sorting based on value counts for each age group
plt.figure(figsize=(10,6))
ax = sns.countplot(x='Age Group', data =store, hue='Gender', palette='Set2', order=age_counts)
for datalabels in ax.containers:
    ax.bar_label(datalabels)
```

```python
##age group generating the highest revenue
age_group_df = store.groupby(['Age Group'],as_index =False)['Amount'].sum()
age_group_df
```

```python
age_group_df['Revenue_percent'] = (age_group_df['Amount']/total_revenue)*100
age_group_df
```

```python
age_group_df.reset_index(inplace=True)
```

```python
plt.figure(figsize=(10,6))
ax= sns.barplot(x='Age Group', y='Revenue_percent',data =age_group_df)
for datalabels in ax.containers:
    ax.bar_label(datalabels,fmt='%.2f')
```

```python
##Another way of showing revenue percent generated by different age groups along with the gender 
age_and_gender =store.groupby(['Age Group','Gender'], as_index =False)['Amount'].sum()
age_and_gender
```

```python
age_and_gender['Revenue percent']=(age_and_gender['Amount']/total_revenue)*100
age_and_gender
```

```python
plt.figure(figsize=(10,6))
ax = sns.barplot(x='Age Group',y='Revenue percent', data=age_and_gender, hue='Gender', palette='Set2')
for datalabels in ax.containers:
    ax.bar_label(datalabels, fmt='%.1f')
```

##### Age groups between 26-35 and 36-45 generate the highest amount of revenue. Amongst these age groups women contribute higher, around 15%-30%

### Revenue Insightes based on Zones, States 

```python
#Distribution of Amount in different Zones
plt.figure(figsize=(10,4))
sns.scatterplot(x='Amount',y='Zone', data=store, color='green')
```

#### Southern and Central Zones have more number of orders that exceed 22,000

```python
#calculate the total revenue generated by each zone
zone_df= store.groupby('Zone', as_index=False)['Amount'].sum()
zone_df.sort_values(by='Amount', ascending=False, inplace=True)
zone_df
```

```python
# Show top 5 states that generate highest revenue and show which zone they belong to
state_df = store.groupby(['State','Zone'], as_index=False)['Amount'].sum()
state_df.sort_values(by='Amount', ascending=False, inplace=True)
top5_states = state_df.head()
```

```python
sns.set(rc={'figure.figsize':(15,5)})
ax= sns.barplot(x='State', y='Amount', data =top5_states,hue='Zone')
for labels in ax.containers:
    ax.bar_label(labels, fmt='%.0f')
plt.show()
```

#### Highest Revenue is genearted by UP from central zone, followed by Maharashtra in the southern zone

## Ocupations which contribute to higher Revenue

```python
#Top 5 occupations that are generating highest revenue
job_df = store.groupby('Occupation', as_index=False)['Amount'].sum()
job_df
```

```python
job_df.sort_values(by='Amount', ascending=False, inplace =True)
```

```python
top5_job =job_df.head()
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.barplot(x='Occupation',y='Amount',data =top5_job,palette='flare')
for labels in ax.containers:
    ax.bar_label(labels, fmt='%.0f')
```

#### Top 5 types of occupations that are generating the highest revenue are IT sector, Healthcare, Aviation, Banking and Govt

### Insights on what products are sold often

```python
#Calculate top 5 products that are sold the most, count the number of orders 
product_df = store.groupby(['Product_Category'], as_index=False).aggregate({'Amount':sum,'Orders':sum})
```

```python
orders_df = product_df.sort_values(by='Orders', ascending=False)
amount_df = product_df.sort_values(by='Amount', ascending=False)
```

```python
top5_orders = orders_df.head()
top5_revenue =amount_df.head()
```

```python
ax = sns.barplot(x='Product_Category', y='Orders', data =top5_orders)
for datalabel in ax.containers:
    ax.bar_label(datalabel)
```

#### Cloathing& Apparel are the most ordered products from the store

```python
ax = sns.barplot(x='Product_Category', y='Amount', data =top5_revenue)
for datalabel in ax.containers:
    ax.bar_label(datalabel)
```

#### Although most orderd product is clothing, food generates highest amount of revenue followed by Clothing and apparel

## Insights

The 26-35 age group, particularly women who work in IT sector, Heathcare and aviation represent a prime target market for food, clothing, and apparel products. Tailored marketing campaigns targeting this demographic in Central and Southern zones could yield substantial returns
